```python
frameIndex = 0
fps = math.ceil(video.fps)
fig = plt.figure()
fig.set_figheight(60)
fig.set_figwidth(60)
        
for f in video.iter_frames():
    #print(divmod(frameIndex, video.fps * 5))
    # every 5 seconds take a image
    div = divmod(frameIndex, fps * 5)
    if div[1] == 0:

        imageIndex = div[0] + 1
        plt.subplot(10, 1, imageIndex)

        img = Image.fromarray(f, 'RGB')
        plt.imshow(img)
        
    frameIndex += 1
    
```
